# Descriptive Statistics

Descriptive statistics is a branch of statistics that involves summarizing and organizing data to describe its main features. 

Unlike inferential statistics, which draw conclusions about a population from a sample, descriptive statistics focus purely on presenting the data in a meaningful way.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mtcars = np.random.randint(15, 201, 50)
mtcars
```

```python
# Mean is the average of the data

mtcars.mean()
```

```python
# Median is the value at the middle of the data, half data is below and half is above

np.median(mtcars)
```

```python
norm_data = pd.DataFrame(np.random.normal(size=100000))

norm_data.plot(kind= "density", figsize=(10, 10))

plt.vlines(norm_data.mean(), ymin=0, ymax=0.4, linewidth=5.0 )

plt.vlines(norm_data.median(), ymin=0, ymax=0.4, linewidth=2.0, color="red" )
```

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Create a random dataset
data = {
    'age': np.random.randint(18, 65, size=100),  # Random ages between 18 and 65
    'income': np.random.normal(50000, 15000, size=100),  # Normally distributed income with mean 50k, std dev 15k
    'hours_worked_per_week': np.random.randint(20, 60, size=100),  # Random hours worked between 20 and 60
    'education_years': np.random.randint(10, 20, size=100)  # Random education years between 10 and 20
}

# Convert to a pandas DataFrame
df = pd.DataFrame(data)

# Perform descriptive statistics
descriptive_stats = df.describe()

# Print the descriptive statistics
print(descriptive_stats)
```

```python
age_stats = df['age'].describe()
print("Descriptive Statistics for Age:")
print(age_stats)
```

```python
hours_worked_stats = df['hours_worked_per_week'].describe()
print("\nDescriptive Statistics for Hours Worked per Week:")
print(hours_worked_stats)
```

**Skewness** measures the asymmetry of a distribution. A perfectly symmetrical distribution (e.g., a normal distribution) has a skewness of 0.

Positive Skewness: Data is skewed right (tail on the right side is longer).

Negative Skewness: Data is skewed left (tail on the left side is longer).

**Why we use it?**
To determine if the data is evenly distributed or if it leans toward higher or lower values. Skewness helps understand the shape of the data distribution.

```python
age_skewness = df['age'].skew()
income_skewness = df['income'].skew()

print(age_skewness)
print()
print(income_skewness)
```

**Kurtosis** measures the "tailedness" of a distribution, or how heavy/light the tails of the distribution are compared to a normal distribution.
High Kurtosis: Indicates more extreme outliers (heavy tails).
Low Kurtosis: Indicates fewer outliers (light tails).

**Why we use it?**
Kurtosis helps us understand whether the data is prone to extreme values and provides insight into the distribution's shape.

```python
age_kurtosis = df['age'].kurtosis()
income_kurtosis = df['income'].kurtosis()

print(age_kurtosis)
print()
print(income_kurtosis)
```

**Quartiles** divide the data into four equal parts:
Q1 (25th Percentile): 25% of the data falls below this value.
Q2 (50th Percentile / Median): 50% of the data falls below this value.
Q3 (75th Percentile): 75% of the data falls below this value.

**Why we use them?**
Quartiles are used to describe the spread of the data and help identify the central tendency (Q2) and the spread or variability in the data (Q1 to Q3).

```python
age_quartiles = df['age'].quantile([0.25, 0.50, 0.75])
income_quartiles = df['income'].quantile([0.25, 0.50, 0.75])

print("Age Quartiles:\n", age_quartiles)
print()
print("Income Quartiles:\n",income_quartiles)
```

**Percentiles** indicate the relative standing of a value in a dataset. For example, the 90th percentile indicates that 90% of the data is below that value.

**Why we use them?**
Percentiles are helpful when we want to understand the distribution of the data in finer detail than quartiles, particularly when looking for extreme values or specific thresholds (e.g., top 10%).

```python
age_percentiles = df['age'].quantile([0.10, 0.90])
income_percentiles = df['income'].quantile([0.10, 0.90])

print("Age Percentiles:\n", age_percentiles)
print()
print("Income Percentiles:\n",income_percentiles)
```
